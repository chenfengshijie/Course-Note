```python
from matplotlib import pyplot as plt
import random
```

```python
class RandomWalk:
    def __init__(self) -> None:
        self.x_path = [0]
        self.y_path = [0]
        self.num_points = 5000

    def walk(self):
        i = 0
        while i < self.num_points - 1:
            dx = random.choice([-1,1])*random.choice([0,1,2,3,4,5])
            dy = random.choice([-1,1])*random.choice([0,1,2,3,4,5])
            if dx == 0 and dy == 0:
                continue
            self.x_path.append(self.x_path[-1]+dx)
            self.y_path.append(self.y_path[-1]+dy)
            i += 1
        return
    
    def plot_map(self):
        # origin map
        # plt.scatter(self.x_path,self.y_path)
        # plt.show()
        # set color        
        plt.axis("off")
        point_num = list(range(self.num_points))
        plt.scatter(self.x_path,self.y_path,c=point_num,cmap=plt.cm.Blues,s=15)
        # set axis invisible


        plt.show()

test = RandomWalk()
test.walk()
test.plot_map()
```
